```python
# importing the image module from matplotlib library

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# loading an image through matplotlib.image module

img = mpimg.imread('/content/JhonWick.jpg')
```

```python
type(img)
```

```python
print(img.shape)
# 3 for 3 chaneels Red green blue 3 matrices
```

```python
print(img)
```

```python
# displaying the image from numpy array

img_plot = plt.imshow(img)
plt.show()
```

```python
#resizing image using PILOW library
from PIL import Image
img=Image.open('/content/JhonWick.jpg')
img_resized=img.resize((250,250))
img_resized.show()#can not do directly
```

```python
img_resized.save('JhonWick_resized.jpg')
```

```python
img_res=mpimg.imread('/content/JhonWick_resized.jpg')
img_Res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

**Converting RGB images to Grayscale image using OpenCV**

```python
# importing OpenCV library
import cv2#Cv:Computer vision
```

```python
img=cv2.imread('/content/JhonWick.jpg')
```

```python
type(img)#converted to numpy array
```

```python
img.shape
```

```python
grayscale_image=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

cv2.imshow() will display the image. But this will not be allowed in Google Colab.

**from google.colab.patches import cv2_imshow**

```python
# displaying the image
from google.colab.patches import cv2_imshow
cv2_imshow(grayscale_image)
```

```python
# saving the grayscale image
cv2.imwrite('JhonWick_grayscale_image.jpg', grayscale_image)
```
